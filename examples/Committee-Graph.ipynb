# Creating the embeddings

```python
import consentiment

# possible by default: laptop14, rest14, hotel22
DATASET = 'hotel22'
MODEL = 'hotel22-bert-base-multilingual-cased-e2eabsa'
EMBEDDING_MODEL = 'bert-base-multilingual-cased'
ROOT = '/content/consentiment/data/'
```

```python
test = consentiment.utils.read_df(ROOT+DATASET+'/test.xlsx', clean_empty=False)
test = test.filter(regex=MODEL+'|review|annotation')
test = test[test.columns.drop(list(test.filter(regex='committee')))]
display(test.head())
```

```python
consentiment.utils.create_embeddings(test, text='review')
```

```python
test.head()
```

```python
G = consentiment.create_graph_from_df(test, text="review", embedding_mi=0.9, distance_threshold=0.1)
```

```python
consentiment.regularize(G)
```

```python
test['graph-committee'] = consentiment.graph_committee_to_data(G, test, text="review")
test
```

```python
spans = consentiment.eval.df_to_spans(test, text="review")
metrics = consentiment.eval.evaluate(spans)
consentiment.eval.metrics_to_df(metrics)
```
