```python
DATASET = 'hotel22'

MODEL = 'hotel22-bert-base-multilingual-cased-e2eabsa'
```

```python
def clean_extractions(l):
    # removes duplicated tuples, when the model extracts the same aspect in
    # different parts of the text

    # print(l)
    counts = {}
    for tup in l:
        counts[tup] = counts.get(tup, 0) + 1
    # print(counts)
    unique = {}
    for tup in counts:
        if tup[0] not in unique:
            unique[tup[0]] = (tup[1], counts[tup])
        elif counts[(tup[0], unique[tup[0]][0])] < counts[tup]:
            unique[tup[0]] = (tup[1], counts[tup])
    # print(unique)
    final = []
    for key in unique:
        final.append((key, unique[key][0]))
    # print(final)
    # print()
    return final
```

```python
import pandas as pd
from ast import literal_eval

ROOT = '/content/consentiment/data/'
test = pd.read_excel(ROOT+DATASET+'/test.xlsx')

test = test.rename(columns={'text':'review'})
test = test.filter(regex=MODEL+'|review')
test = test[test.columns.drop(list(test.filter(regex='committee')))]

for c in test.columns:
    if c == 'review': continue
    print(c)
    test[c] = test[c].apply(literal_eval)
    test[c] = test[c].apply(clean_extractions)
test
```

```python
import pandas as pd
import numpy as np

id_to_sent = {0:'POS', 1:'NEU', 2:'NEG'}
sent_to_id = {}
for key in id_to_sent: sent_to_id[id_to_sent[key]] = key

def merge_similar_aspects(aspect_list):
    # It will basically merge to the smallest common existing aspect
    # o processador + processador = processador
    # café da manhã + café = café
    merged_aspects = {}
    for aspect in aspect_list:
        merged = False
        for key in list(merged_aspects):
            if key in aspect:
                merged_aspects[aspect] = key
                merged = True
            elif aspect in key:
                merged_aspects[key] = aspect
                merged_aspects[aspect] = aspect
                for key2 in list(merged_aspects):
                    if merged_aspects[key2] == key:
                        merged_aspects[key2] = aspect
                merged = True
        if not merged:
            merged_aspects[aspect] = aspect
    return merged_aspects

def solve_conflicts(counts):
    # will get the most voted sentiment
    final = []
    for key in counts:
        max_value = np.max(counts[key])
        argmax_positions = np.where(counts[key] == max_value)[0]

        if len(argmax_positions) == 1:
            final.append((key, id_to_sent[argmax_positions[0]]))
        else:
            # when there are sentiments tied
            final_id = 1
            if len(argmax_positions) == 2:
                # pos + neu = pos
                if argmax_positions[0] == 0 and argmax_positions[1] == 1:
                    final_id = 0
                # neg + neu = neg
                elif argmax_positions[0] == 1 and argmax_positions[1] == 2:
                    final_id = 2
                # otherwise = neu (pos == neg and pos == neu == neg)
            final.append((key, id_to_sent[final_id]))
    return final

def committee_vote(df):
    committee_ans = []
    for index, row in df.iterrows():
        # if index != 80: continue

        # Getting all aspects in a row
        aspect_list = []
        for col in row.index[2:]:
            aspect_list += [tup[0] for tup in row[col]]

        # Getting merged aspects -> food: food, the food: food

        merged_aspects = merge_similar_aspects(aspect_list)

        # Counting tuples frequency (aspect, sent): count
        counts = {}
        for col in row.index[2:]:
            for (aspect, sentiment) in row[col]:
                sent_c = counts.get(merged_aspects[aspect], [0, 0, 0])
                sent_c[sent_to_id[sentiment]] += 1
                counts[merged_aspects[aspect]] = sent_c

        counts_solved = solve_conflicts(counts)

        # print(aspect_list)
        # print(merged_aspects)
        # print(counts)
        # print(counts_solved)
        # print()

        committee_ans.append(counts_solved)

    return committee_ans

votes = committee_vote(test)
votes
```
